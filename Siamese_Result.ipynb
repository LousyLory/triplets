## Siamese Results With Map Background Texture

#### Every target image is being compared with 31 test images. Out of 31 test images, 20% of the images are positive i.e. contains same word as that of the target image. 

```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

# import system things
import tensorflow as tf
import numpy as np
import os
import pickle
from util import get_test_results
import cv2

from data_generator_for_siamese import ImageDataGenerator
from datetime import datetime
from tensorflow.contrib.data import Iterator
from tensorflow.contrib.data import Dataset
from helper import get_dataset
import siamese
from matplotlib import pyplot as plt
import glob
import matplotlib.image as mpimg
import re
%matplotlib inline

def merge_image_with_background(input_image):
    kernel = np.ones((3,3), np.uint8)
    image = cv2.dilate(input_image, kernel, iterations=1)
    
    h,w = image.shape[0], image.shape[1]
    
    if h>227 and w>227:
        image = cv2.resize(image, (227, 227))
    elif h>227:
        image = cv2.resize(image, (w, 227))
    elif w>227:
        image = cv2.resize(image, (227, h))
    h,w = image.shape[0], image.shape[1]
    img = np.zeros((227, 227, 3), np.uint8)
    img[:, :, :] = 255
    centerX, centerY = int(h/2), int(w/2)

    shiftX, shiftY = int(227/2) - centerX, int(227/2) - centerY
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if image[i, j] == 255:
                #print(i+shiftX, j+shiftY)
                img[i+shiftX, j+shiftY, :] = 0
    return img
            
    
# prepare data and tf.session
sess = tf.InteractiveSession()
num_classes = 999
batch_size = 31
# setup siamese network
siamese_model = siamese.siamese_network(batch_size);
train_step = tf.train.AdamOptimizer(0.001).minimize(siamese_model.loss)
saver = tf.train.Saver()

train_file = 'word_train.txt'
val_file = 'word_val.txt'
# Place data loading and preprocessing on the cpu
with tf.device('/cpu:0'):
    tr_data = ImageDataGenerator(train_file,
                                 mode='training',
                                 batch_size=batch_size,
                                 num_classes=num_classes,
                                 shuffle=True)

sess.run(tf.global_variables_initializer())
# start training

#siamese_model.load_initial_weights(sess)
num_epochs = 300
new = False

saver.restore(sess, 'trained_models/model.ckpt')

embed = None
flag = False

f = open("../test_map/list.txt",'rb')
lines = f.readlines()
f.close()
words = []
for line in lines:
    words.append(re.sub(r'[^a-zA-Z0-9]','',line))

images = glob.glob("../test_map/resized/*.jpg")
for img in images:
    color_image = mpimg.imread(img)
    gray = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
    ret,thresh1 = cv2.threshold(gray,127,255,cv2.THRESH_BINARY_INV)
    
    img = merge_image_with_background(thresh1)
    
    #plt.imshow(img)
    #plt.show()
    
    #batch_x1, color_images, test_words = tr_data.get_batch_from_word_list(words, img, color_image)
    batch_x1, test_words = tr_data.get_batch_from_word_list(words, img, color_image)
    word_list = batch_x1
    result_vectors = siamese_model.o1.eval({siamese_model.x1: batch_x1})
    result = get_test_results(test_words, result_vectors)
    print("Original Image")
    #plt.imshow(color_images[0])
    plt.imshow(batch_x1[0])
    plt.show()
    print("*"*20)
    print("Top 5 matching images")
    print("*"*20)
    for i in range(5):
        plt.imshow(batch_x1[result[i][2]])
        print("Distance:", result[i][1])
        plt.show()
```
