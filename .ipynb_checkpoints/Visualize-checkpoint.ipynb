```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as sio
#import tensorflow as tf
```

```python
from vis_files import visualize_matches
f = open('splits/test_files.txt', 'r')
A = f.readlines()
f.close()
A = [x.strip('\n') for x in A]
print(A)

match_scores = 0
for filename in A:
    Im = np.load('D:\\word\\original_images\\nopad\\original_images_nopad_'+filename+'.tiff.npy')
    words = np.load('D:\\word\\synthetic_word_images\\nobg_nopad\\sythetic_word_images_nobg_nopad_'+filename+'.npy')
    
    #### compute the matches
    match, closest_word = visualize_matches(Im, words)
    match_score = np.sum(match) / float(len(words))
    print(match_score)
    match_scores += match_score
pass
```

```python
#### average score:
average_score = match_scores / len(A)
print(average_score)
```
